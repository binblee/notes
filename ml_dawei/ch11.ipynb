聚类

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
digits = load_digits()
x, y = digits.data, digits.target
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=33, test_size=0.25)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
def cluster_ars(n):
    scores = []
    num_range = list(range(2, n+1))
    for i in num_range:
        kms = KMeans(i, n_init=10)
        kms.fit(x_train)
        predicts = kms.predict(x_test)
        scores.append(metrics.adjusted_rand_score(y_test,predicts))
    plt.figure()
    plt.plot(num_range, scores, '*-')
    plt.xlabel('# of clusters')
    plt.ylabel('Adjusted Rand Score')
    plt.plot()

def cluster_ss(n):
    sc_scores = []
    num_range = list(range(2, n+1))
    for i in num_range:
        kms = KMeans(i, n_init=10)
        kms.fit(x_train)
        sc_score=metrics.silhouette_score(x_train,kms.fit(x_train).labels_,metric='euclidean')
        sc_scores.append(sc_score)
        
    plt.figure()
    plt.plot(num_range, sc_scores, '*-')
    plt.xlabel('# of clusters')
    plt.ylabel('Silhouette Coefficient Score')
    plt.plot()
```

```python
cluster_ars(20)
```

```python
cluster_ss(20)
```
