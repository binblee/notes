```python
import cv2
from plt_utils import plt_imgshow
```

```python
def detect_corners(filename):
    img = cv2.imread(filename)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dst = cv2.cornerHarris(gray, 2, 23, 0.04)
    img[dst > 0.01*dst.max()] = [0, 0, 255]
    return img
```

```python
corners = detect_corners('../images/chess_board.png')
plt_imgshow(corners)
```

```python
corners = detect_corners('../images/track.png')
plt_imgshow(corners)
```
