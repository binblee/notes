```python
import cv2
import numpy as np
from scipy import ndimage
from plt_utils import plt_imgshow
```

```python
kernel_3x3 = np.array([[-1,-1,-1],
                       [-1,8,-1],
                       [-1,-1,-1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1,  1,  2,  1, -1],
                       [-1,  2,  4,  2, -1],
                       [-1,  1,  2,  1, -1],
                       [-1, -1, -1, -1, -1]])


img = cv2.imread('../images/statue_small.jpg', 0)


k3 = ndimage.convolve(img, kernel_3x3)
k5 = ndimage.convolve(img, kernel_5x5)
blured = cv2.GaussianBlur(img, (17,17), 0)
g_hpf = img - blured
```

```python
plt_imgshow(img)
```

```python
plt_imgshow(k3)
```

```python
plt_imgshow(k5)
```

```python
plt_imgshow(blured)
```

```python
plt_imgshow(g_hpf)
```
