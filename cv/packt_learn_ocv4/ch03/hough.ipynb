```python
import cv2
import numpy as np
from plt_utils import plt_imgshow
```

```python
img = cv2.imread('../images/houghlines5.jpg')
plt_imgshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 120)
minLineLength = 20
maxLineGap = 5
lines = cv2.HoughLinesP(edges, 1, np.pi/180.0, 20, minLineLength, maxLineGap)
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
plt_imgshow(img)
```

```python
planets = cv2.imread('../images/planet_glow.jpg')
plt_imgshow(planets)
```

```python
gray = cv2.medianBlur(cv2.cvtColor(planets, cv2.COLOR_BGR2GRAY), 5)
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 120, 
                           param1=100, param2=30, minRadius=0, maxRadius=0)
circles = np.uint16(np.around(circles))
print(len(circles), circles)
```

```python
for c in circles[0,:]:
    cv2.circle(planets, (c[0], c[1]), c[2], (0, 255, 0), 2)
    # draw center
    cv2.circle(planets, (c[0], c[1]), 2, (0, 0, 255), 3)
plt_imgshow(planets)
```
